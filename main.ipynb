## Section 3: Feature Engineering

### Create new variables

Total Charges per Month: Divide the total charges by the tenure to get an average monthly spending. This could be more indicative of the customer's spending behavior.

Number of Services: Create a feature that represents the total number of services the customer has signed up for. This could capture the overall engagement of the customer.

Scale numeric features like monthly charges and total charges to a similar scale. This is important for algorithms that rely on distance measures, such as Support Vector Machines or k-Nearest Neighbors.


- Number of services

```python
df_binary = df.copy()

df_binary["PhoneService"] = df_binary["PhoneService"].map({'No': 0, 'Yes': 1})
df_binary["MultipleLines"] = df_binary["MultipleLines"].map({'No': 0, 'Yes': 1, "No phone service": 0})
df_binary["InternetService"] = 1
df_binary["OnlineSecurity"] = df_binary["OnlineSecurity"].map({'No': 0, 'Yes': 1, "No phone service": 0})
df_binary["OnlineBackup"] = df_binary["OnlineBackup"].map({'No': 0, 'Yes': 1, "No phone service": 0})
df_binary["DeviceProtection"] = df_binary["DeviceProtection"].map({'No': 0, 'Yes': 1, "No phone service": 0})
df_binary["TechSupport"] = df_binary["TechSupport"].map({'No': 0, 'Yes': 1, "No phone service": 0})
df_binary["StreamingTV"] = df_binary["StreamingTV"].map({'No': 0, 'Yes': 1, "No phone service": 0})
df_binary["StreamingMovies"] = df_binary["StreamingMovies"].map({'No': 0, 'Yes': 1,"No phone service": 0})

columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
                   'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
                   'StreamingMovies']

# Create the "Total_services" variable by summing up the one-hot-encoded columns
df_binary['Total_services'] = df_binary[columns].sum(axis=1)
df_binary.head()
```

- Service_ease

```python
df_binary["PaperlessBilling"] = df_binary["PaperlessBilling"].map({'No': 0, 'Yes': 1})
df_binary["PaymentMethod"] = df_binary["PaymentMethod"].map({'Electronic check': 0, 'Mailed check': 0, 
                                                             'Bank transfer (automatic)': 1, 'Credit card (automatic)': 1})

columns = ["PaperlessBilling", "PaymentMethod"]

df_binary['Service_ease'] = df_binary[columns].sum(axis=1) # 2 = High, 1 = Moderate, 0 = Low
df_binary["Service_ease"] = df_binary["Service_ease"].map({0: "Low", 1: "Moderate", 2: "High"})

df_binary.head()
```

- Total Charges per Month 



```python
df_binary['av_monthly_spending'] = df_binary["TotalCharges"]/df_binary['tenure']
```

## Section 4: Modeling

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
from sklearn.metrics import RocCurveDisplay
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.metrics import precision_recall_curve
```

```python
X = df_resampled_manual.drop(columns='Churn')
y = df_resampled_manual['Churn']

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 2) 

x_fit, x_val, y_fit, y_val = train_test_split(x_train, y_train, random_state = 2)
```

```python
def model(classifier,x_fit,y_fit,x_val,y_val):
    
    classifier.fit(x_fit,y_fit)
    prediction = classifier.predict(x_val)
    cv = RepeatedStratifiedKFold(n_splits = 10,n_repeats = 3,random_state = 1)
    print("Cross Validation Score : ",'{0:.2%}'.format(cross_val_score(classifier,x_fit,y_fit,cv = cv,scoring = 'roc_auc').mean()))
    print("ROC_AUC Score : ",'{0:.2%}'.format(roc_auc_score(y_val,prediction)))
    RocCurveDisplay.from_estimator(classifier, x_val,y_val)
    plt.title('ROC_AUC_Plot')
    plt.show()

def model_evaluation(classifier,x_test,y_test):
    
    # Confusion Matrix
    cm = confusion_matrix(y_test,classifier.predict(x_test))
    names = ['True Neg','False Pos','False Neg','True Pos']
    counts = [value for value in cm.flatten()]
    percentages = ['{0:.2%}'.format(value) for value in cm.flatten()/np.sum(cm)]
    labels = [f'{v1}\n{v2}\n{v3}' for v1, v2, v3 in zip(names,counts,percentages)]
    labels = np.asarray(labels).reshape(2,2)
    sns.heatmap(cm,annot = labels,cmap = 'Blues',fmt ='')
    
    # Classification Report
    print(classification_report(y_test,classifier.predict(x_test)))
```

```python
from xgboost import XGBClassifier
classifier_xgb = XGBClassifier(learning_rate= 0.01,
                               max_depth = 10,
                               n_estimators = 1500)
```

```python
model(classifier_xgb,x_fit,y_fit,x_val,y_val)
```

```python
model_evaluation(classifier_xgb,x_train,y_train)
```

```python
model_evaluation(classifier_xgb,x_test,y_test)
```
